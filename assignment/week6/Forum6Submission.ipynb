```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm,datasets
from sklearn.svm import  SVC
import numpy as np
from sklearn.metrics import accuracy_score

df = pd.read_csv('aftercleaning.csv')
df.info()

#143745 - mean
#classification - by using the mean
a = df['Reported number of people receiving ART'].mean()
df['label']=np.where(df['Reported number of people receiving ART']>a,'good','bad')
df

```

```python
#cutting the data only up to 100
X=df.iloc[0:76,[1,4,7,9]]
print(X)
y=df.loc[0:75,'label']
print(y)

```

```python
#if c is inreased more accurate; c=1 accuracy 1
clf = svm.SVC(kernel='rbf', random_state=1, gamma=1, C=1)

#train the model
clf.fit(X, y)

#just use the same data to predict for now
y_pred=clf.predict(X)


print(accuracy_score(y, y_pred))
```
